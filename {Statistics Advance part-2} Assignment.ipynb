**1. What is hypothesis testing in statistics?**

Hypothesis Testing is a statistical method used to decide whether the data behind a claim or assumption supports that claim or not.

- Testing a claim: Hypothesis testing means checking a claim or assumption using data.

- Two hypotheses are formed:
    - Null Hypothesis (H‚ÇÄ): The default claim we assume is true.
    - Alternative Hypothesis (H‚ÇÅ): The claim we want to prove (that H‚ÇÄ is false).

- Sample data is used: We use data from a small group (sample) to check if the claim is valid for the whole population.

- We use P-value and significance level: A P-value is calculated and compared with a significance level (usually 0.05) to make a decision.

- We make a decision ‚Äì accept or reject:
    - If the P-value is less than 0.05, we reject the null hypothesis.
    - Otherwise, we accept (or fail to reject) it.

**2. What is the null hypothesis, and how does it differ from the alternative hypothesis?**

Null vs Alternative Hypothesis:

- The Null Hypothesis (H‚ÇÄ) is the default assumption: It assumes ‚Äúthere is no effect‚Äù or ‚Äúnothing has changed.‚Äù For example ‚Äì the medicine has no effect.

- The Alternative Hypothesis (H‚ÇÅ) is the new claim: It suggests that ‚Äúsomething has changed‚Äù or the null hypothesis is incorrect. For example ‚Äì the medicine has an effect.

- Hypothesis testing checks H‚ÇÄ using data: We test whether the sample data gives us enough reason to reject H‚ÇÄ.

- We reject H‚ÇÄ when the P-value is small: If the data shows strong evidence against H‚ÇÄ, we reject it and support H‚ÇÅ.

- H‚ÇÄ and H‚ÇÅ are always opposites: If H‚ÇÄ says "population mean = 50", then H‚ÇÅ will say "population mean ‚â† 50", or maybe "> 50" or "< 50".

**3.  What is the significance level in hypothesis testing, and why is it important?**

Significance Level:

- The significance level, denoted by Œ± (alpha), is a value used in hypothesis testing that defines the probability of rejecting the null hypothesis when it is actually true. 

- It represents the risk of making a Type error. 

- For example, if we choose a significance level of 0.05, it means we are allowing a 5% chance that we may reject the null hypothesis even though it is true.


Importance of significance level:

- It determines the reliability of the test results.

- It controls the chances of making an error in research.

- It helps us understand how strong our conclusions are based on the data.

- A smaller significance level means we are being more strict about rejecting the null hypothesis, which increases the confidence in our results.

**4. What does a P-value represent in hypothesis testing?**

The P-value is a statistical value that shows the probability of obtaining results as extreme as the observed data, assuming that the null hypothesis is true.

In simple words, the P-value tells us how likely our sample results are under the assumption that the null hypothesis is correct.

- If the P-value is very small (like less than 0.05), we reject the null hypothesis.

- If the P-value is large, we do not reject the null hypothesis.

- It helps us decide whether the result is statistically significant or not.

- Thus, the smaller the P-value, the stronger the evidence against the null hypothesis.

**5.  How do you interpret the P-value in hypothesis testing?**

- The interpretation of the P-value is based on comparing it with the chosen significance level (Œ±), usually 0.05.

- If P-value < Œ± (e.g., 0.05): We reject the null hypothesis.

- It means the result is statistically significant.

- The observed data is unlikely under the null hypothesis.

If P-value ‚â• Œ±:

- We do not reject the null hypothesis.

- It means there is not enough evidence to say the result is significant.

- The data supports the null hypothesis.

Example:

- If Œ± = 0.05 and the P-value = 0.02, then we interpret that the result is statistically significant and we reject the null hypothesis.

**6. What are Type 1 and Type 2 errors in hypothesis testing?**

In hypothesis testing, there are two types of possible errors: Type 1 and Type 2 errors.

Type 1 Error:

- This occurs when we reject the null hypothesis even though it is actually true.

- It is also called a False Positive.

Example:

- Convicting an innocent person in court.The probability of this error is denoted by Œ± (alpha).

Type 2 Error:

- This happens when we fail to reject the null hypothesis even though it is actually false.

- It is called a False Negative.

- The probability of this error is denoted by Œ≤ (beta).

Example:

- Letting a guilty person go free in court.

**12.  What is the T-test, and how is it used in hypothesis testing?**

**1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and 
interpret the results?**

Suppose the mean of a population is 100, and its standard deviation is 15.

Now we take a sample, and the mean is 105.

So the question arises:

Is this 105 just luck or is it actually something different?

The Z-test answers this.

```python
from scipy import stats
import numpy as np

population_mean = 100
population_std = 15

sample_data = [102, 108, 105, 107, 106, 104, 103, 109, 101, 106]

sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print('sample_mean:', sample_mean)
print('Z-statistics:', z_statistic)
print('P-value:', p_value)

alpha = 0.05
if p_value < alpha:
    print("Result: Statistically significant ‚Üí Reject the null hypothesis")
else:
    print("Result: Not statistically significant ‚Üí Fail to reject the null hypothesis")
```

**4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot.**

```python
import matplotlib.pyplot as plt

sample_data = [102, 98, 100, 105, 97, 99, 101, 96, 104, 103]

population_mean = 100
population_std = 5  
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Z-test calculation
z_stat = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Decision rule
alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)  # two-tailed critical value (e.g. ¬±1.96 for 95% confidence)

if abs(z_stat) > z_critical:
    print("Result: Statistically significant ‚Üí Reject the null hypothesis")
else:
    print("Result: Not statistically significant ‚Üí Fail to reject the null hypothesis")

# Step 6: Plot the decision region
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x < -z_critical) | (x > z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z-stat = {z_stat:.2f}')
plt.title("Two-tailed Z-Test Visualization")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

**5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define required variables
mu0 = 0      # Mean under H‚ÇÄ
mu1 = 1      # Mean under H‚ÇÅ
sigma = 1    # Standard deviation
alpha = 0.05

# Z critical values
z_crit = norm.ppf(1 - alpha/2)

# X-axis values
x = np.linspace(-4, 6, 1000)

# PDFs for H‚ÇÄ and H‚ÇÅ
y0 = norm.pdf(x, mu0, sigma)
y1 = norm.pdf(x, mu1, sigma)

# Plot distributions
plt.plot(x, y0, label='H‚ÇÄ (Null)', color='blue')
plt.plot(x, y1, label='H‚ÇÅ (Alternate)', color='green')

# Type I error (Œ±)
plt.fill_between(x, y0, where=(x < mu0 - z_crit*sigma) | (x > mu0 + z_crit*sigma), color='red', alpha=0.3, label='Type I Error (Œ±)')

# Type II error (Œ≤)
plt.fill_between(x, y1, where=(x > mu0 - z_crit*sigma) & (x < mu0 + z_crit*sigma), color='orange', alpha=0.3, label='Type II Error (Œ≤)')

plt.axvline(mu0 - z_crit*sigma, linestyle='--', color='gray')
plt.axvline(mu0 + z_crit*sigma, linestyle='--', color='gray')

plt.title("Type I and Type II Error Visualization")
plt.xlabel("Test Statistic")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

**6. Write a Python program to perform an independent T-test and interpret the results.**

**9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.**

```python
import numpy as np
from scipy.stats import t

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = np.std(sample, ddof=1) / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n-1)
    margin_of_error = t_critical * std_err
    lower = mean - margin_of_error
    upper = mean + margin_of_error
    return (lower, upper)
```

```python
sample_data = [50, 52, 48, 49, 51, 53, 47]
ci = confidence_interval(sample_data, confidence=0.95)
print("95% Confidence Interval:", ci)
```

Explanation:
- A confidence interval gives a range where we expect the true population mean to fall, based on our sample.
- For a 95% CI, it means:
    - ‚ÄúWe are 95% confident that the true mean lies between the lower and upper bounds.‚Äù
- The wider the interval, the more uncertainty in the estimate.
- The larger the sample, the narrower the interval (more accurate).

```python
import numpy as np
from scipy.stats import t

def margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_dev = np.std(sample, ddof=1)
    std_err = std_dev / np.sqrt(n)
    t_critical = t.ppf((1 + confidence) / 2, df=n - 1)
    moe = t_critical * std_err
    return moe

# Sample data
sample_data = [48, 50, 52, 49, 51, 53, 47]

# Confidence level
confidence_level = 0.95

# Calculate Margin of Error
moe = margin_of_error(sample_data, confidence=confidence_level)
print(f"Margin of Error at {int(confidence_level * 100)}% confidence level: {moe:.2f}")
```

**11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.**

Where:
- P(H‚à£E) = Posterior (probability of hypothesis H given evidence E)
- P(H) = Prior (initial belief about H)
- P(E‚à£H) = Likelihood (probability of evidence given H)
- (E) = Marginal likelihood (total probability of evidence)

```python
def bayes_theorem(prior_H, likelihood_E_given_H, prior_not_H, likelihood_E_given_not_H):
    # Total probability of evidence
    P_E = (likelihood_E_given_H * prior_H) + (likelihood_E_given_not_H * prior_not_H)
    
    # Bayes' Theorem
    posterior = (likelihood_E_given_H * prior_H) / P_E
    return posterior
```

Example:

```python
# Prior probability of having the disease
P_D = 0.001

# Probability of testing positive if diseased
P_Pos_D = 0.99

# Prior probability of NOT having the disease
P_not_D = 0.999

# Probability of testing positive if not diseased (false positive rate)
P_Pos_not_D = 0.05

# Apply Bayes' Theorem
posterior = bayes_theorem(P_D, P_Pos_D, P_not_D, P_Pos_not_D)

print(f"Probability of having the disease after testing positive: {posterior:.4f}")
```

Explanation of the process:

Prior belief (P_D) is the chance someone has the disease before testing.

Likelihood is how accurate the test is in both positive and false positive cases.

Bayes' Theorem combines your prior with the evidence (test result) to give the posterior, or updated probability.

You‚Äôll see a low result (like ~2%) even though the test is 99% accurate ‚Äî because the disease is very rare.

**12. Perform a Chi-square test for independence between two categorical variables in Python.**

Suppose we want to check if Gender and Product Preference are related.

```python
import pandas as pd
import scipy.stats as stats

#Create a contingency table (cross-tab)
# Example data: Gender vs Product Preference
data = {
    'Product A': [30, 20],
    'Product B': [10, 25]
}
index = ['Male', 'Female']
table = pd.DataFrame(data, index=index)

print("Contingency Table:")
print(table)

#Perform Chi-square test
chi2, p, dof, expected = stats.chi2_contingency(table)

#Show results
print("\nChi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=index, columns=table.columns))

#Interpret result
alpha = 0.05
if p < alpha:
    print("Conclusion: Variables are dependent ‚Üí Reject the null hypothesis.")
else:
    print("Conclusion: Variables are independent ‚Üí Fail to reject the null hypothesis.")
```

**16. Implement an F-test using Python to compare the variances of two random samples.**

```python
import numpy as np
from scipy.stats import f

# Generate two random samples
np.random.seed(42)
sample1 = np.random.normal(loc=50, scale=10, size=30)  # mean=50, std=10
sample2 = np.random.normal(loc=50, scale=15, size=30)  # mean=50, std=15

# Calculate variances
var1 = np.var(sample1, ddof=1)  # sample variance (unbiased)
var2 = np.var(sample2, ddof=1)

# Ensure var1 is the larger variance
F = var1 / var2 if var1 > var2 else var2 / var1

# Degrees of freedom
df1 = len(sample1) - 1
df2 = len(sample2) - 1

# Calculate p-value (two-tailed)
p_value = 2 * min(f.cdf(F, df1, df2), 1 - f.cdf(F, df1, df2))

# Output
print("Variance 1:", round(var1, 4))
print("Variance 2:", round(var2, 4))
print("F-statistic:", round(F, 4))
print("p-value:", round(p_value, 4))

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: Variances are not significantly different.")
```

Notes:
- Always use ddof=1 to calculate sample variance.

- F-test is sensitive to non-normality ‚Äî it assumes both samples are from normal distributions.

- You can also use scipy.stats.levene or bartlett for robustness checks.

**19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.**

```python
import numpy as np
from scipy.stats import shapiro, levene

def check_anova_assumptions(*groups):
    print("Checking ANOVA assumptions:\n")
    for i, group in enumerate(groups, start=1):
        _, p = shapiro(group)
        print(f"Group {i} Shapiro-Wilk p = {p:.4f} ‚Üí", "Normal" if p > 0.05 else "Not Normal")
    _, p = levene(*groups)
    print(f"\nLevene‚Äôs Test p = {p:.4f} ‚Üí", "Equal variances" if p > 0.05 else "Unequal variances")
    print("\nIndependence is assumed.")

group1 = [85, 90, 88, 75, 95]
group2 = [70, 65, 78, 72, 68]
group3 = [90, 88, 94, 89, 92]

check_anova_assumptions(group1, group2, group3)
```

**24. Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.**

```python
import numpy as np
from scipy.stats import chi2

# Step 1: Simulate sample data
np.random.seed(0)
sample = np.random.normal(loc=50, scale=5, size=30)  # mean=50, std=5

# Step 2: Sample stats
n = len(sample)
sample_var = np.var(sample, ddof=1)

# Step 3: Hypothesized population variance
pop_var_hypothesis = 25  # std^2 = 5^2

# Step 4: Calculate chi-square test statistic
chi2_stat = (n - 1) * sample_var / pop_var_hypothesis

# Step 5: Calculate p-value for two-tailed test
df = n - 1
p_value = 2 * min(chi2.cdf(chi2_stat, df), 1 - chi2.cdf(chi2_stat, df))

# Step 6: Show results
print("Sample Variance:", round(sample_var, 4))
print("Chi-square Statistic:", round(chi2_stat, 4))
print("Degrees of Freedom:", df)
print("P-value:", round(p_value, 4))

# Step 7: Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H‚ÇÄ: Population variance is significantly different.")
else:
    print("Fail to reject H‚ÇÄ: No significant difference in variance.")
```

**26. Implement an F-test for comparing the variances of two datasets, then interpret and visualize the results.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import f

# Step 1: Simulate two datasets
np.random.seed(0)
data1 = np.random.normal(loc=50, scale=10, size=100)
data2 = np.random.normal(loc=50, scale=15, size=100)

# Step 2: Calculate sample variances
var1 = np.var(data1, ddof=1)
var2 = np.var(data2, ddof=1)

# Step 3: Determine F-statistic (larger variance in numerator)
F = max(var1, var2) / min(var1, var2)
df1 = len(data1) - 1
df2 = len(data2) - 1

# Step 4: Calculate two-tailed p-value
p_value = 2 * min(f.cdf(F, df1, df2), 1 - f.cdf(F, df1, df2))

# Step 5: Print results
print("Variance 1:", round(var1, 2))
print("Variance 2:", round(var2, 2))
print("F-statistic:", round(F, 4))
print("P-value:", round(p_value, 4))
if p_value < 0.05:
    print("Reject H‚ÇÄ: Variances are significantly different.")
else:
    print("Fail to reject H‚ÇÄ: No significant difference in variances.")

# Step 6: Visualization
plt.figure(figsize=(8, 5))
plt.hist(data1, bins=20, alpha=0.6, label='Dataset 1', color='skyblue')
plt.hist(data2, bins=20, alpha=0.6, label='Dataset 2', color='orange')
plt.title('Histogram: Comparing Variances')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
